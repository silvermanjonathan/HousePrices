```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data_train.csv')

df.rename(columns = {'Unnamed: 0': 'index'}, inplace = True)

df.set_index('index', inplace = True)
```

```python
set([type(i) for i in [df[i][0] for i in df.columns]])
```

```python
'''
According to this output we see that data types are sting and numpy.floats, numpy integers, and strings.
strings are not suitable for data analysis. Let's grab the columns whose values are numbers. 
'''

column_names_dict = dict(enumerate(df.columns)) 
column_types_dict = dict(enumerate([type(i) for i in [df[i][0] for i in df.columns]]))

depend_vars_iloc = []

for k,v in column_types_dict.items():
    if v != type('hi'):
        depend_vars_iloc.append(k)
depend_vars_iloc
```

```python
''' We will now make subplots checking for linear relationships between the depend_vars and the independant vars'''



fig, axes = plt.subplots(5, 4, figsize=(16,16))
fig.suptitle('scatter plots of independant variables')


indep_var = df['price']
sns.scatterplot(ax=axes[0,0], x = df.iloc[:,0], y = indep_var)
sns.scatterplot(ax=axes[0,1], x = df.iloc[:,2], y = indep_var)
sns.scatterplot(ax=axes[0,2], x = df.iloc[:,3], y = indep_var)
sns.scatterplot(ax=axes[0,3], x = df.iloc[:,4], y = indep_var)

sns.scatterplot(ax=axes[1,0], x = df.iloc[:,5], y = indep_var)
sns.scatterplot(ax=axes[1,1], x = df.iloc[:,6], y = indep_var)
sns.scatterplot(ax=axes[1,2], x = df.iloc[:,7], y = indep_var)
sns.scatterplot(ax=axes[1,3], x = df.iloc[:,8], y = indep_var)

sns.scatterplot(ax=axes[2,0], x = df.iloc[:,9], y = indep_var)
sns.scatterplot(ax=axes[2,1], x = df.iloc[:,10], y = indep_var)
sns.scatterplot(ax=axes[2,2], x = df.iloc[:,11], y = indep_var)
sns.scatterplot(ax=axes[2,3], x = df.iloc[:,12], y = indep_var)

sns.scatterplot(ax=axes[3,0], x = df.iloc[:,13], y = indep_var)
sns.scatterplot(ax=axes[3,1], x = df.iloc[:,14], y = indep_var)
sns.scatterplot(ax=axes[3,2], x = df.iloc[:,15], y = indep_var)
sns.scatterplot(ax=axes[3,3], x = df.iloc[:,16], y = indep_var)

sns.scatterplot(ax=axes[4,0], x = df.iloc[:,17], y = indep_var)
sns.scatterplot(ax=axes[4,1], x = df.iloc[:,18], y = indep_var)
sns.scatterplot(ax=axes[4,2], x = df.iloc[:,19], y = indep_var)
sns.scatterplot(ax=axes[4,3], x = df.iloc[:,20], y = indep_var)



fig.tight_layout()
plt.show()
```
